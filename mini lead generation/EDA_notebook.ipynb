
# Exploratory Data Analysis (EDA) of LinkedIn Company Data

## Introduction:

LinkedIn is a popular platform for professionals and companies to network, share insights, and connect with potential clients or partners. In this notebook, we perform an exploratory data analysis (EDA) on company data scraped from LinkedIn. The dataset contains information about various companies, including their industry, location, and other relevant details.

The purpose of this analysis is to gain insights into the distribution of industries and locations among the companies in the dataset. By exploring the data, we aim to uncover patterns, trends, and potential leads that could be valuable for business development or market research purposes.

We will start by loading the cleaned dataset and examining its contents. Then, we'll conduct visualizations and statistical analyses to explore the distribution of industries and locations. Finally, we'll provide interpretations of the findings and suggest potential areas for further analysis.



##Step1:

We begin by importing the necessary libraries for the EDA that are **pandas**,**matplotlib.pyplot**, **seaborn**, and **wordcloud**.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
```

##Step2:
The second step involves read the cleaned dataset.

```python
cleaned_data = pd.read_csv('linkedin.csv')
summary_statistics = cleaned_data.describe(include='object')
print(summary_statistics)
```

##Step3:
It consists of building a countplot shows the distribution of companies across different industries. It appears that the "Consulting for Firms" (Conseil aux entreprises) industry is the most common among the companies in the dataset.

```python
plt.figure(figsize=(12, 6))
sns.countplot(y='Industry', data=cleaned_data, order=cleaned_data['Industry'].value_counts().index)
plt.title('Distribution of Industries')
plt.xlabel('Count')
plt.ylabel('Industry')
plt.show()
```

##Step4:
The countplot displays the top 10 locations of companies. It seems that the majority of companies are located in the United States, particularly in New York and Boston, MA.

```python
plt.figure(figsize=(12, 6))
sns.countplot(y='Location', data=cleaned_data, order=cleaned_data['Location'].value_counts().index[:10])
plt.title('Top 10 Locations of Companies')
plt.xlabel('Count')
plt.ylabel('Location')
plt.show()
```

##Step5:
The Word Cloud of industry names gives us a wider and more clear view on what are the major specializations of Consulting companies.

```python
industry_names = ' '.join(cleaned_data['Industry'])
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(industry_names)
plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.title('Word Cloud of Industries')
plt.axis('off')
plt.show()
```

#Conclusion:

In this exploratory data analysis (EDA) of LinkedIn company data, we gained valuable insights into the distribution of industries and locations among the companies in the dataset. Here are the key findings and takeaways:

- **Distribution of Industries**: The countplot revealed that the "Consulting for firms" industry is the most common among the companies in the dataset, followed by "Information Technology & Services" and "HR". This suggests a high representation of consulting firms on LinkedIn.

- **Distribution of Locations**: The countplot of locations showed that the majority of companies are located in the United States, particularly in cities like New York,and Boston followed by India in Gurgaon. This indicates a concentration of companies in major business hubs in these two countries which we can add Canada to the list too.

- **Word Cloud of Industries**: The word cloud visually represented the frequency of different industries in the dataset. The larger the word, the more common the industry. This visualization confirmed the prevalence of industries like consulting for firms, IT, and recrutment.

Overall, this analysis provides valuable insights for business development, market research, and networking purposes.

By leveraging the rich data available on LinkedIn, businesses and professionals can make informed decisions, identify potential leads, and uncover opportunities for collaboration and growth.
